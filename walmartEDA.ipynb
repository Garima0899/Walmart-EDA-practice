```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dg = pd.read_csv(r"C:\Users\hp\Downloads\walmart-sales-dataset-of-45stores (1).csv") 
dg
```

```python
# Sales Analysis Over Time
# How do weekly sales vary over time for each store?



```

```python
dg.info()
```

```python
weekly_sales =dg.groupby(['Store'])['Weekly_Sales'].sum().sort_values(ascending= False)
weekly_sales
```

```python
weekly_sales.plot(kind= 'bar')   # from pandas
```

```python
# Are there any noticeable trends or seasonal patterns in sales?
```

```python
dg['Date'].info()
```

```python
dg['Month'] = dg['Date'].str[3:5]
dg
```

```python
seasonal_trend =dg.groupby(['Month'])[['Weekly_Sales']].mean()
seasonal_trend
```

```python
sns.lineplot(data = seasonal_trend, x='Month', y='Weekly_Sales')  # sales increased in the month of oct,nov,dec
```

```python
# Holiday Impact

 How do holiday weeks impact weekly sales compared to non-holiday weeks?
# Which holidays have the most significant impact on sales?
```

```python
holiday = dg.groupby(['Holiday_Flag'])[['Weekly_Sales']].mean().reset_index()
holiday
```

```python
# Which holidays have the most significant impact on sales?   .. sales are more in the holiday week 
holiday.plot(kind= 'bar')
```

```python
 Which stores have the highest and lowest average weekly sales?
```

```python
store = dg.groupby(['Store'])[['Weekly_Sales']].mean().sort_values(by = 'Weekly_Sales', ascending = False)
store
```

```python
print('Highest sales : ',store.head(1))
print('Lowest sales : ',store.tail(1))
```

```python
How does the performance of different stores compare in terms of sales over time?
```

```python
store.plot(kind = 'bar')
```

```python
Is there a correlation between temperature and weekly sales?
```

```python
corre = dg['Temperature'].corr(dg['Weekly_Sales'])  # neagative correlation means temp increases then weekly sales decreases
corre
```

```python
dg
```

```python
How do sales vary with different temperature ranges?
```

```python
dg['Temp_Range'] = np.where(dg['Temperature'] <= 30, '<=30',
                      np.where(dg['Temperature'] <= 40, '30-40',
                      np.where(dg['Temperature'] <= 50, '40-50',
                      np.where(dg['Temperature'] <= 60, '50-60',
                      np.where(dg['Temperature'] <= 70, '60-70',
                      np.where(dg['Temperature'] <= 80, '70-80',
                      np.where(dg['Temperature'] <= 90, '80-90', '>90')))))))
dg
```

```python
sns.boxplot(data = dg, x = 'Temp_Range', y = 'Weekly_Sales')
```
